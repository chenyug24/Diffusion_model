```python
def load_model_and_extract_att_mask(model_path, model, device):
    """
    get attention mask
    """
    checkpoint = torch.load(model_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    att_mask_pro = model.att_mask_pro

    return att_mask_pro

def generate_adj_matrix(att_mask_pro, threshold=0.5):
    """
    根据 attention mask 的概率生成邻接矩阵，使用阈值来二值化。
    """
    adj_matrix = (att_mask_pro > threshold).int()
    return adj_matrix.cpu().numpy()
```

```python
from google.colab import drive
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
from dagma.linear import DagmaLinear

def compute_metrics(B_true, B_est):
    d = B_true.shape[0]
    mask_no_diag = 1 - np.eye(d)
    pred = (np.abs(B_est) > 1e-4).astype(int)
    true = (B_true != 0).astype(int)

    TP = np.sum((true == 1) & (pred == 1) & (mask_no_diag == 1))
    FP = np.sum((true == 0) & (pred == 1) & (mask_no_diag == 1))
    TN = np.sum((true == 0) & (pred == 0) & (mask_no_diag == 1))
    FN = np.sum((true == 1) & (pred == 0) & (mask_no_diag == 1))

    fdr = FP / (TP + FP) if (TP + FP) > 0 else 0.0
    tpr = TP / (TP + FN) if (TP + FN) > 0 else 0.0
    fpr = FP / (FP + TN) if (FP + TN) > 0 else 0.0
    nnz = np.sum(pred * mask_no_diag)

    return {'fdr': fdr, 'tpr': tpr, 'fpr': fpr, 'nnz': nnz}

# Mount Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

samples_list = range(5)
results = []
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_nodes_list = [20,30,50,80,100]

# --- NOISE SETTING ---
ADD_NOISE = False  # Disabled as requested
NOISE_STD = 0.0

print(f"Starting Comparison (Gaussian Noise Added: {ADD_NOISE})...")

for num_nodes in num_nodes_list:
    for sample in samples_list:
        print(f"--- Nodes {num_nodes}, Sample {sample} ---")
        # --- Path Definition ---
        data_path = f'/content/drive/MyDrive/Path_squared_new/data_{sample}_{num_nodes}.pt'
        model_path = f'/content/drive/MyDrive/Path_squared_new/checkpoint_d{num_nodes}_exp{sample}.pth'

        # 1. Load Ground Truth & Data
        t_start = time.time()
        try:
            data_all = torch.load(data_path, weights_only=False)
        except Exception as e:
            print(f"Skipping {data_path}: {e}")
            continue

        graph_adj = data_all['graph_adj'] # True Graph

        if isinstance(data_all['data'], torch.Tensor):
            data_values = data_all['data'].cpu().numpy()
        else:
            data_values = data_all['data']

        # --- ADD GAUSSIAN NOISE ---
        if ADD_NOISE:
            noise = np.random.normal(0, NOISE_STD, data_values.shape)
            data_values = data_values + noise
            # Normalize again to keep DAGMA stable
            data_values = (data_values - np.mean(data_values, axis=0)) / np.std(data_values, axis=0)

        print(f"  Data Loaded ({time.time() - t_start:.2f}s)")

        # 2. Get Diffusion Result
        t_load = time.time()
        try:
            checkpoint = torch.load(model_path, map_location=device)
            if 'model_state_dict' in checkpoint:
                state_dict = checkpoint['model_state_dict']
            else:
                state_dict = checkpoint

            if 'att_mask_pro' in state_dict:
                att_mask_logits = state_dict['att_mask_pro']
                probs = torch.softmax(att_mask_logits, dim=-1)
                edge_probs = probs[:, :, 1]

                adj_diffusion = (edge_probs > 0.5).int().cpu().numpy()
            else:
                print(f"  'att_mask_pro' not found in {model_path}")
                adj_diffusion = np.zeros((num_nodes, num_nodes))

        except Exception as e:
            print(f"  Error loading model {model_path}: {e}")
            adj_diffusion = np.zeros((num_nodes, num_nodes))
        print(f"  Model/Mask Extracted ({time.time() - t_load:.2f}s)")


        # 3. Get DAGMA Result
        t_dagma = time.time()
        dagma = DagmaLinear(loss_type='l2')
        W_dagma = dagma.fit(data_values, lambda1=0.02)
        adj_dagma = (np.abs(W_dagma) > 0).astype(int)
        print(f"  DAGMA Finished ({time.time() - t_dagma:.2f}s)")

        # 4. Compute Metrics
        m_diff = compute_metrics(graph_adj, adj_diffusion)
        m_dagma = compute_metrics(graph_adj, adj_dagma)

        results.append({'sample': sample, 'num_nodes': num_nodes, 'model': 'Diffusion', **m_diff})
        results.append({'sample': sample, 'num_nodes': num_nodes, 'model': 'DAGMA', **m_dagma})

        print(f"  Result: Diff FDR={m_diff['fdr']:.2f}, DAGMA FDR={m_dagma['fdr']:.2f}")

# ==========================================
# 4. Plotting
# ==========================================
if results:
    df = pd.DataFrame(results)
    metrics = ['fdr', 'tpr', 'fpr', 'nnz']
    fig, axes = plt.subplots(1, 4, figsize=(24, 5))

    for i, m in enumerate(metrics):
        sns.barplot(x='num_nodes', y=m, hue='model', data=df, ax=axes[i], capsize=.1, palette="Set2")
        axes[i].set_title(m.upper())

    plt.tight_layout()
    plt.show()

    print(df.groupby(['num_nodes', 'model'])[metrics].mean())
else:
    print("No results to plot.")
```
